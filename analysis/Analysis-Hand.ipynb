```python
## Tokenize sentence and find the NP that 'some' is modifying
import stanza
nlp = stanza.Pipeline('en')
import re

fwords = ['of', 'the', ',', 'uh', 'um']

def find_np(sent):
    doc = nlp(sent)
    if len(doc.sentences) > 1:
        return ''
    if len(list(re.finditer('some', sent.lower()))) > 1:
        return ''
    some_ind = [word.text.lower() for word in doc.sentences[0].words].index('some') + 1
    np = ' '.join([word.text.lower() for word in doc.sentences[0].words if (word.head==some_ind and 
                                                                           word.text.lower() not in fwords)])
    return np
```

```python
uniq = data[['Item', 'Sentence']].drop_duplicates('Item')
uniq = uniq.set_index('Item')
print(uniq.shape)
uniq['NP'] = uniq['Sentence'].apply(lambda x: find_np(x))
uniq = uniq[uniq['NP']!='']
print(uniq.shape)
```
